```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.optimizers import Adam
from model.segmentation import UNet
from util.function import *
from util.metric import *
def main():
    data_path = './traindata/' #train and validation data path
    model_save_path = "./weight/" #set the path to save weight file
    makedir(model_save_path)
    #load train data
    imgs_train = np.load(data_path+"train_image.npy").astype(np.float32)
    imgs_mask_train = np.load(data_path+"train_mask.npy").astype(np.float32)
    #load model
    learning_rate=1.0e-5
    model = UNet()
    model.compile(optimizer=Adam(lr=learning_rate), loss=dice_coef_loss, metrics=[dice_coef])    
    model_path = os.path.join(model_save_path, "nodule_segmentation_model.h5")
    model_checkpoint = ModelCheckpoint (model_path, monitor='loss', save_best_only=True)
    
    batch_size=2 #set batch size
    epochs=100 #set epochs
    print("start train")
    train_history =model.fit(imgs_train, imgs_mask_train, batch_size=batch_size, epochs=epochs, verbose=1, shuffle=True,callbacks=[model_checkpoint])
    model.save_weights(os.path.join(model_save_path, "nodule_segmentation_model_last.h5")) #save weight at last epoch
if __name__ == "__main__":
    main()
```
