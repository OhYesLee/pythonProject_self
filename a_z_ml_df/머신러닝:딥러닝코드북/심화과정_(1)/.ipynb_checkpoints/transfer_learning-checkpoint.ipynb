# Transfer learning

### 기존에 학습한 pretrained 모델을 로드한 후, 학습을 진행하는 transfer learning을 배워본다.

#### (본 코드에서는 동일한 task인 mnist classification 모델을 로드했지만 본래 transfer learning은 동일하지는 않지만 연관된 task에서 학습한 pretrained 모델을 로드한다.)

```python
##tenssorflow==2.3 환경에서 구현 
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist

(mnist_train, mnist_train_label), (mnist_test, mnist_test_label) = mnist.load_data() #mnist data road
#normalization 적용
mnist_train=mnist_train/255
mnist_test=mnist_test/255

mnist_train=tf.cast(mnist_train, tf.float32)
mnist_train=np.expand_dims(mnist_train, -1)
mnist_test=tf.cast(mnist_test, tf.float32)
mnist_test=np.expand_dims(mnist_test, -1)
```

```python
#기존에 학습한 전체 model 로드
from tensorflow.keras.models import load_model
model =load_model("./mnist_classification_model.h5")

#transfer learning 진행
model.fit(mnist_train, mnist_train_label, epochs=100)
print("train finished")
```

```python
#최종 train accuracy 측정
metrics=model.evaluate(mnist_train, mnist_train_label, verbose=2)
print("Train accuracy:", metrics[1])
```

```python
#test accuracy 측정
metrics=model.evaluate(mnist_test, mnist_test_label, verbose=2)
print("Test accuracy:", metrics[1])
```
