# Visualize Heatmap

### 기존에 학습한 mnist pretrained 모델을 로드한 후, gradcam을 이용하여 mnist 테스트 데이터 이미지에 heatmap을 그려본다.

#### (gradcam 생성 방법은 create_heatmap.py 파일 참고)

```python
##tenssorflow==2.3 환경에서 구현 
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist
from random import *
import cv2
import matplotlib.pyplot as plt

(mnist_train, mnist_train_label), (mnist_test, mnist_test_label) = mnist.load_data() #mnist data load

# normalization 적용
mnist_test=mnist_test/255

mnist_test=tf.cast(mnist_test, tf.float32)
mnist_test=np.expand_dims(mnist_test, -1)
```

```python
# mnist test 데이터에서 임의의 이미지 샘플링
sample_index = randint(0, len(mnist_test)-1)  
img=mnist_test[sample_index]
# mnist 이미지 숫자 확인을 위해 샘플링한 이미지 시각화
plt.imshow(img)
```

```python
# 기존에 학습한 전체 model 로드
from tensorflow.keras.models import load_model
model =load_model("./mnist_classification_model.h5")
model.summary()
```

```python
from create_heatmap import gradcam
layer_number=-4 # 전체 모델에서 관찰하려는 layer 선택. (이 경우, mnist_classification_model.h5 모델에서 뒤에서 4번째 layer인 dense layer 선택)
image_number=1 #위에서 시각화한 mnist 이미지를 확인하고 mnist 이미지 숫자 입력
# gradcam 이미지 생성
gradcam_img=gradcam(img,image_number,model,layer_number)
```

```python
# heatmap 생성 
gradcam_img = cv2.resize(gradcam_img, (224, 224))
resize_img = cv2.resize(img, (224, 224))
# gradcam 이미지만 시각화
plt.figure()
plt.imshow(gradcam_img, alpha=0.5)
# mnist 이미지에 gradcam 이미지를 입혀서 시각화
plt.figure()
plt.imshow(resize_img)
plt.imshow(gradcam_img, alpha=0.5)
```
